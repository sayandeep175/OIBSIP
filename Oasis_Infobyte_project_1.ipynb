```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as pt

import seaborn as sea

import plotly.express as px
```

```python
train = pd.read_csv("Unemployment in India.csv")

train = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(train.head())
```

```python
print(train.isnull().sum())
```

```python
train.columns= ["States","Date", "Frequency",

"Estimated Unemployment Rate",

"Estimated Employed", "Estimated Labour Participation Rate",

"Region","longitude","latitude"]
```

```python
pt.style.use('seaborn-whitegrid')

pt.figure(figsize=(8,6))

sea.heatmap(train.corr())

pt.show()
```

```python
train.columns= ["States", "Date", "Frequency",

"Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region", "longitude","latitude"]

pt.title("Indian Unemployment")

sea.histplot(x="Estimated Employed", hue="Region", data=train)

pt.show()
```

```python
pt.figure(figsize=(8, 6))


pt.title("Indian Unemployment")

sea.histplot(x="Estimated Unemployment Rate", hue="Region", data=train)

pt.show()
```

```python
Unemploment = train[["Region","Estimated Unemployment Rate"]]

fig = px.sunburst(Unemploment, path=["Region"], values="Estimated Unemployment Rate", width=500, height=500, color_continuous_scale="RdY1Gn", title="Unemployment Rate in India")

fig.show()
```
